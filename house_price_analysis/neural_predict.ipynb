```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pickle

with open("train_test_split_price.pkl", "rb") as f:
    X_train_scaled, X_test_scaled, y_train_boxcox, y_test_boxcox, fitted_lambda,y_shift = pickle.load(f)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

from sklearn.metrics import r2_score, mean_squared_error
from scipy.special import inv_boxcox
from numpy import sqrt
```

```python
model = Sequential([
    Dense(32, activation='relu', input_shape=(8,)),
    Dropout(0.2),
    Dense(16, activation='relu'),
    Dense(8, activation='relu'),
    Dense(1, activation='linear')
])
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mse','mae'])
```

```python
model.summary()
```

```python
history = model.fit(X_train_scaled, y_train_boxcox,
                    epochs=200,
                    batch_size=32)
```

```python
y_pred_nn = model.predict(X_test_scaled).flatten()
```

```python
y_pred_org_nn = inv_boxcox(y_pred_nn, fitted_lambda) - y_shift
y_test_org_nn = inv_boxcox(y_test_boxcox, fitted_lambda) - y_shift
```

```python
r2_score_nn = r2_score(y_test_org_nn, y_pred_org_nn)
mse_nn = mean_squared_error(y_test_org_nn, y_pred_org_nn)
rmse_nn = sqrt(mse_nn)

print(f"Keras NN R2 Score : {r2_score_nn:.4f}")
print(f"Keras NN MSE      : {mse_nn:.4f}")
print(f"Keras NN RMSE     : {rmse_nn:.4f}")
```
